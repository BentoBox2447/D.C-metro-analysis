# Cleaning the Data


I want to break down the full date column into a year, month and day column, and rename the day column to a full_date column

```python
data = data.rename(columns = {'day' : 'full_date'}) 
```

```python
# this will break down our full_date column and make 3 new columns ('year', 'month', 'day')
data['year']=[d.split('-')[0] for d in data.full_date]
data['month']=[d.split('-')[1] for d in data.full_date]
data['day']=[d.split('-')[2]for d in data.full_date]
```
